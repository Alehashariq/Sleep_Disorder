```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Sample Centroids (for two clusters)
C1 = np.array([2.0, 3.0, 4.0])
C2 = np.array([5.0, 6.0, 7.0])

# 1. Euclidean Distance
def euclidean_distance(c1, c2):
    return np.linalg.norm(c1 - c2)

# 2. Manhattan Distance
def manhattan_distance(c1, c2):
    return np.sum(np.abs(c1 - c2))

# 3. Cosine Similarity / Distance
def cosine_distance(c1, c2):
    similarity = np.dot(c1, c2) / (np.linalg.norm(c1) * np.linalg.norm(c2))
    return 1 - similarity

# 4. Mahalanobis Distance
def mahalanobis_distance(c1, c2, data):
    covariance_matrix = np.cov(data.T)
    inv_cov_matrix = np.linalg.inv(covariance_matrix)
    diff = c1 - c2
    return np.sqrt(np.dot(np.dot(diff.T, inv_cov_matrix), diff))

# 5. Jaccard Distance (for binary data)
def jaccard_distance(c1, c2):
    intersection = np.sum(np.minimum(c1, c2))
    union = np.sum(np.maximum(c1, c2))
    return 1 - intersection / union

# Example Data for Mahalanobis Distance
sample_data = np.array([[2, 3, 4], [5, 6, 7], [2, 5, 6], [3, 4, 7]])

# Calculating distances
euclidean = euclidean_distance(C1, C2)
manhattan = manhattan_distance(C1, C2)
cosine = cosine_distance(C1, C2)
mahalanobis = mahalanobis_distance(C1, C2, sample_data)

# Plotting distances
methods = ['Euclidean', 'Manhattan', 'Cosine', 'Mahalanobis', 'Jaccard']
distances = [euclidean, manhattan, cosine, mahalanobis, jaccard_distance(np.array([1, 0, 1, 1, 0]), np.array([1, 1, 0, 1, 0]))]

plt.figure(figsize=(10, 5))
plt.bar(methods, distances, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Comparison of Different Centroid Distance Measures')
plt.xlabel('Distance Measure')
plt.ylabel('Distance Value')
plt.show()

# 3D Plot of Centroids
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(C1[0], C1[1], C1[2], color='blue', s=100, label='Centroid 1')
ax.scatter(C2[0], C2[1], C2[2], color='red', s=100, label='Centroid 2')

ax.set_title('3D Visualization of Centroids')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```

```python
import numpy as np
from scipy.spatial import distance
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from mpl_toolkits.mplot3d import Axes3D

# Sample Centroids (for two clusters)
C1 = np.array([2.0, 3.0, 4.0])
C2 = np.array([5.0, 6.0, 7.0])

# 1. Euclidean Distance
def euclidean_distance(c1, c2):
    return np.linalg.norm(c1 - c2)

# 2. Manhattan Distance
def manhattan_distance(c1, c2):
    return np.sum(np.abs(c1 - c2))

# 3. Cosine Similarity / Distance
def cosine_distance(c1, c2):
    similarity = np.dot(c1, c2) / (np.linalg.norm(c1) * np.linalg.norm(c2))
    return 1 - similarity

# 4. Mahalanobis Distance
def mahalanobis_distance(c1, c2, data):
    covariance_matrix = np.cov(data.T)
    inv_cov_matrix = np.linalg.inv(covariance_matrix)
    diff = c1 - c2
    return np.sqrt(np.dot(np.dot(diff.T, inv_cov_matrix), diff))

# 5. Jaccard Distance (for binary data)
def jaccard_distance(c1, c2):
    intersection = np.sum(np.minimum(c1, c2))
    union = np.sum(np.maximum(c1, c2))
    return 1 - intersection / union

# 6. Similarity Index (1 - Distance)
def similarity_index(distance_value):
    return 1 - distance_value

# Example Data for Mahalanobis Distance
sample_data = np.array([[2, 3, 4], [5, 6, 7], [2, 5, 6], [3, 4, 7]])

# Calculating distances
euclidean = euclidean_distance(C1, C2)
manhattan = manhattan_distance(C1, C2)
cosine = cosine_distance(C1, C2)
mahalanobis = mahalanobis_distance(C1, C2, sample_data)
jaccard = jaccard_distance(np.array([1, 0, 1, 1, 0]), np.array([1, 1, 0, 1, 0]))

# Calculating Similarity Indices
similarities = [similarity_index(euclidean), similarity_index(manhattan), similarity_index(cosine), similarity_index(mahalanobis), similarity_index(jaccard)]

# Plotting distances
methods = ['Euclidean', 'Manhattan', 'Cosine', 'Mahalanobis', 'Jaccard']
distances = [euclidean, manhattan, cosine, mahalanobis, jaccard]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(methods, distances, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Comparison of Different Centroid Distance Measures')
plt.xlabel('Distance Measure')
plt.ylabel('Distance Value')

# Plotting similarity indices
plt.subplot(1, 2, 2)
plt.bar(methods, similarities, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title('Similarity Indices (1 - Distance)')
plt.xlabel('Similarity Measure')
plt.ylabel('Similarity Index')

plt.tight_layout()
plt.show()

# 3D Plot of Centroids
fig = plt.figure(figsize=(7, 7))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(C1[0], C1[1], C1[2], color='blue', s=100, label='Centroid 1')
ax.scatter(C2[0], C2[1], C2[2], color='red', s=100, label='Centroid 2')

ax.set_title('3D Visualization of Centroids')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.legend()

plt.show()
```
