```python
import dash
from dash import dcc, html, Input, Output
import pandas as pd
import plotly.express as px

# Sample Data
data = {
    "country": ["Afghanistan"] * 7 + ["India"] * 7,
    "date": [1960, 1961, 1962, 1963, 1964, 1965, 1966] * 2,
    "Population": [8622466.0, 8790140.0, 8969047.0, 9157465.0, 9355514.0, 9565147.0, 9783147.0] +
                  [447150000.0, 453100000.0, 459300000.0, 465800000.0, 472600000.0, 479600000.0, 486800000.0],
    "Net Migration": [2606.0, 6109.0, 7016.0, 6681.0, 7079.0, 10695.0, 3563.0] +
                     [-18000.0, -17000.0, -16000.0, -15000.0, -14000.0, -13000.0, -12000.0]
}

df = pd.DataFrame(data)

# Initialize Dash App
app = dash.Dash(__name__)

app.layout = html.Div([
    html.H1("Population and Net Migration Dashboard", style={'textAlign': 'center'}),
    
    # Dropdown for country selection
    html.Div([
        html.Label("Select Country:"),
        dcc.Dropdown(
            id='country-dropdown',
            options=[{'label': country, 'value': country} for country in df['country'].unique()],
            value='Afghanistan'  # Default value
        )
    ], style={'width': '50%', 'margin': 'auto'}),
    
    # Graphs
    html.Div([
        dcc.Graph(id='line-plot'),
        dcc.Graph(id='scatter-plot')
    ])
])

# Callbacks for interactive plots
@app.callback(
    [Output('line-plot', 'figure'),
     Output('scatter-plot', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_country):
    # Filter data based on selected country
    filtered_df = df[df['country'] == selected_country]

    # Line plot: Population over time
    line_fig = px.line(
        filtered_df,
        x='date',
        y='Population',
        title=f'Population Over Time: {selected_country}',
        labels={'date': 'Year', 'Population': 'Total Population'}
    )
    
    # Scatter plot: Net Migration vs Population
    scatter_fig = px.scatter(
        filtered_df,
        x='Population',
        y='Net Migration',
        title=f'Net Migration vs Total Population: {selected_country}',
        labels={'Population': 'Total Population', 'Net Migration': 'Net Migration'},
        size='Net Migration',  # Bubble size
        color='date',  # Color by year
        hover_name='date'
    )
    
    return line_fig, scatter_fig

if __name__ == '__main__':
    app.run_server(debug=True)
```
