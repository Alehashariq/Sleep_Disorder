
t-Test

Choosing between a Z-test and a T-test for hypothesis testing depends primarily on two factors: the sample size and whether the population standard deviation is known.
Z-test:

    When to Use:
        The population standard deviation is known.
        The sample size is large (commonly, n ≥ 30). With large samples, the sample standard deviation approximates the population standard deviation.
        For proportions (e.g., testing the proportion of success in a sample against a known population proportion).

    Characteristics:
        Based on the Z-distribution, which is a normal distribution as n becomes large.
        More commonly used in quality control and standardization processes.

T-test:

    When to Use:
        The population standard deviation is unknown.
        The sample size is small (typically, n < 30).
        Suitable for cases where the data is approximately normally distributed, especially in small samples.

    Characteristics:
        Based on the T-distribution, which accounts for the additional uncertainty due to the estimation of the population standard deviation from the sample.
        T-distribution becomes closer to the normal distribution as the sample size increases.

General Guidelines:

    Large Samples: With large sample sizes, the T-test and Z-test will give similar results. This is because the T-distribution approaches the normal distribution as the sample size increases.
    Small Samples: When the sample size is small and the population standard deviation is unknown, the T-test is generally the appropriate choice due to its ability to account for the uncertainty in the standard deviation estimate.
    Unknown Population Standard Deviation: Even with large samples, if the population standard deviation is unknown and cannot be reliably estimated, a T-test is usually preferred.

Conclusion:

    Use the Z-test for large sample sizes or when the population standard deviation is known.
    Use the T-test for small sample sizes or when the population standard deviation is unknown.

In practice, the T-test is more commonly used in many research scenarios due to the rarity of knowing the population standard deviation and often dealing with smaller sample sizes.

# 📊 **Why is the t-test Called "Student’s t-test"?**  

## 🔹 **Historical Background**  
The **t-test** is called the **Student’s t-test** because it was developed by a statistician named **William Sealy Gosset**.  

- Gosset worked at **Guinness Brewery** 🍺 in the early 1900s.  
- Guinness had strict policies against employees publishing research.  
- To bypass this, Gosset **published his work under the pseudonym** **"Student"** in 1908.  
- His paper, *"The Probable Error of a Mean"*, introduced the **t-distribution**, which led to the famous **t-test**.  

## 🔹 **Why Was the t-test Created?**  
The **t-test** is useful when:  
✅ The **sample size is small** (*n < 30*).  
✅ The **population variance is unknown**.  
✅ We need to **compare means** and make statistical inferences.  

## 🔹 **Types of t-tests**  
1️⃣ **One-Sample t-test** → Compares a sample mean to a **known** population mean.  
2️⃣ **Independent (Two-Sample) t-test** → Compares means of **two independent groups**.  
3️⃣ **Paired t-test** → Compares means of **the same group** before and after an intervention.  

---

## 💡 **Fun Fact**  
Gosset developed the t-test to **improve beer quality** 🍻 at Guinness by analyzing small batches of ingredients!  




```python
import scipy.stats as stats

# sample data
x = [1, 2, 3, 4, 5]

# known population mean
mu = 4

# perform one sample t-test
t_statistic, p = stats.ttest_1samp(x, mu)

# print results
print('t-statistic:', t_statistic)
print('p-value:', p)

# print the results using if else
if p > 0.05:
    print(f'p-value: {p}, Sample mean is equal to population mean (fail to reject H0)')
else:
    print(f'p-value: {p}, Sample mean is not equal to population mean (reject H0)')
```

```python
# 🔹 Hypotheses for the t-test

#     Null Hypothesis (H₀): There is no significant difference in the average fare between survivors and non-survivors. (Fare did not influence survival)
#     Alternative Hypothesis (H₁): There is a significant difference in the average fare between survivors and non-survivors. (Fare influenced survival)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Drop NaN values for accurate t-test calculations
titanic = titanic.dropna(subset=['fare', 'survived'])

# Define two groups: Survivors and Non-Survivors
survivors = titanic[titanic['survived'] == 1]['fare']
non_survivors = titanic[titanic['survived'] == 0]['fare']

# Perform an Independent t-test (Welch’s t-test since variance may not be equal)
t_stat, p_value = stats.ttest_ind(survivors, non_survivors, equal_var=False)

# Print results with hypothesis context
print("📌 Hypothesis Testing for Titanic Fare and Survival")
print(f"H₀: No significant difference in fare between survivors and non-survivors.")
print(f"H₁: Significant difference in fare between survivors and non-survivors.")
print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.5f}")

# 📈 Visualization: Fare Distribution for Survivors & Non-Survivors
plt.figure(figsize=(10, 5))
sns.kdeplot(survivors, fill=True, label="Survived", color="green", alpha=0.6)
sns.kdeplot(non_survivors, fill=True, label="Did Not Survive", color="red", alpha=0.6)
plt.axvline(np.mean(survivors), color='green', linestyle='dashed', linewidth=2, label="Survivor Mean")
plt.axvline(np.mean(non_survivors), color='red', linestyle='dashed', linewidth=2, label="Non-Survivor Mean")
plt.title("Fare Distribution: Survivors vs. Non-Survivors", fontsize=14)
plt.xlabel("Fare")
plt.ylabel("Density")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

# Given data
data = [11, 24, 22, 18, 16, 32, 38, 48, 58, 44]

# Calculate sample mean
x_bar = np.mean(data)

# Calculate sample standard deviation (s) using (n-1) degrees of freedom
s = np.std(data, ddof=1)

# Calculate standard error (SE)
n = len(data)
SE = s / np.sqrt(n)

# Print results
print(f"Sample Mean (x̄): {x_bar}")
print(f"Sample Standard Deviation (s): {s}")
print(f"Standard Error (SE): {SE}")
```

```python
import pandas as pd
import scipy.stats as stats

# Load Titanic dataset
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Drop NaN values in the 'Age' column
df = df.dropna(subset=['Age'])

# Split data into two groups: Male and Female passengers
male_ages = df[df["Sex"] == "male"]["Age"]
female_ages = df[df["Sex"] == "female"]["Age"]

# Perform independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(male_ages, female_ages, equal_var=False)

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

# Decision making
if p_value > 0.05:
    print("Fail to reject H0: No significant difference in age between male and female passengers")
else:
    print("Reject H0: Significant difference in age between male and female passengers")
```


Test for Homogeniety
1. Levene's Test

```python
import scipy.stats as stats

# Sample data
group1 = [2.3, 3.4, 4.5, 2.3, 3.4]
group2 = [1.2, 2.2, 3.2, 2.2, 2.3]

# Perform Levene's Test
w_stat, p_val = stats.levene(group1, group2)

print("w-statistic:", w_stat)
print("p-value:", p_val)

# print the results based on if else conditions
if p_val > 0.05:
    print(f'p-value: {p_val}, Variances are equal (fail to reject H0)')
else:
    print(f'p-value: {p_val}, Variances are not equal (reject H0)')
```

Analysis of Variance
1. ONe-way ANOVA

```python
import scipy.stats as stats

# Sample data

a = [2.3, 3.4, 4.5, 2.3, 3.4]
b = [11.2, 12.2, 13.2, 12.2, 12.3]
c = [22.3, 23.4, 24.5, 22.3, 23.4]

# perform one-way ANOVA
f_stat, p_val = stats.f_oneway(a, b, c)

# print the results base on if else conditions
if p_val > 0.05:
    print(f'p-value: {p_val}, Means are equal (fail to reject H0)')
else:
    print(f'p-value: {p_val}, Means are not equal (reject H0)')
```
